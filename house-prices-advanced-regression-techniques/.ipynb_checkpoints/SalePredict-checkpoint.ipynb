# Baseline

## Nettoyage 

```python
df = pd.read_csv('train.csv')
```

```python
df
```

```python
X = df.drop(columns=['Id', 'SalePrice'], axis=1)
y = df.SalePrice
```

```python
X.info()
```

```python
a = X.select_dtypes('object').nunique() <= 5
df_select_obj = pd.DataFrame(a).reset_index()
df_better = df_select_obj[df_select_obj[0] == True]
df_good_col = df_better['index']
list_col = list(df_good_col)


```

```python
features_num = list(X.select_dtypes('int64','float64'))
features_num
```

```python
features_cat = list_col
features_cat
```

## Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import Ridge
```

```python
pipe_cat = Pipeline([('impute', SimpleImputer(strategy='most_frequent')),
                     ('encode', OneHotEncoder(handle_unknown='ignore'))])

pipe_num = Pipeline([('impute_num', SimpleImputer()),
                     ('encode_num', MinMaxScaler())])

preprocessing = ColumnTransformer([('pipe_cat', pipe_cat, features_cat),
                                   ('features_num', pipe_num, features_num)])
```

```python
preprocessing
```

```python
model = Pipeline([('prepro', preprocessing), ('ridge', Ridge())])
model.fit(X, y)
```

## GridSearchCV

```python
from sklearn.model_selection import GridSearchCV

param = {
    'ridge__solver':
    ['auto', 'svd', 'cholesky', 'lsqr', 'sparse_cg', 'sag', 'saga']
}

grid = GridSearchCV(model, param, cv=5)
```

```python
grid.fit(X,y)
```

## Score et predict

```python
grid.best_score_
```

```python
new_model = grid.best_estimator_
print(f'Le modele proposÃ© est {new_model} ')
print(f'Son score est de {new_model.score(X,y)}')
```

```python
model.score(X,y)
```

```python
test = pd.read_csv('test.csv')
X_test = test.drop(columns='Id')
```

```python
ypred = new_model.predict(X_test)
ypred
```

### Pipeline

```python
from sklearn.linear_model import LinearRegression


pipe_cat_od = Pipeline([('impute', SimpleImputer(strategy='most_frequent')),
                        ('encode', OrdinalEncoder())])

pipe_cat_ohe = Pipeline([('impute', SimpleImputer(strategy='most_frequent')),
                         ('encode', OneHotEncoder(handle_unknown='ignore'))])


pipe_num = Pipeline([('impute_num', SimpleImputer()),
                     ('encode_num', MinMaxScaler())])

prepro = ColumnTransformer([('pipe_cat_od', pipe_cat_od, ordinal),
                            ('pipe_cat_ohe', pipe_cat_ohe, cat),
                            ('features_num', pipe_num, X_final_num)])
```
